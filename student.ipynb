### Model Creation

Selecting the appropriate regression model for this dataset largely depends on the nature of the problem you're trying to solve.

### Splitting feature and target 

```python
x = df.drop('Performance Index', axis=1)
y = df['Performance Index']
```

### Features : All columns except perfomance index will be used as features.

```python
x
```

### Target : The performance index column will be used as the target variable.

```python
y
```

### Linear Regression

### Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)
```

### Step 1 : import the model

```python
from sklearn.linear_model import LinearRegression
```

### Step 2 : instatiate the model

```python
lr = LinearRegression()
```

### Step 3 : Fit the model to the training data

```python
lr.fit(xtrain,ytrain)
```

### step 4 : make prediction

```python
ypred=lr.predict(xtest)
```

### Model Evaluation

```python
lr.coef_
```

```python
lr.intercept_
```

```python
ypred
```

```python
plt.scatter(ytest,ypred)
```

```python
sns.distplot(ytest-ypred)
```

```python
pd.DataFrame(lr.coef_,index=x.columns,columns=["coefficients"])
```

### model evaluation matrix

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
mae=mean_absolute_error(ytest,ypred)
mse=mean_squared_error(ytest,ypred)
rmse=np.sqrt(mse)
r2=r2_score(ytest,ypred)
print("Mean Absolute Error : ",mae)
print("Mean Squared Error : ",mse)
print("Root Mean Squared Error : ",rmse)
print("r2_score (Accuracy): ",r2)
```

### Plotting the results
### Actual vs Predicted performance index

```python
plt.figure(figsize=(10, 6))
plt.scatter(ytest, ypred, color='blue')
plt.plot([y.min(), y.max()], [y.min(), y.max()], color='red', linestyle='--', linewidth=2)
plt.xlabel('Actual Performance Index')
plt.ylabel('Predicted Performance Index')
plt.title('Actual vs Predicted Performance Index')
plt.show()
```
